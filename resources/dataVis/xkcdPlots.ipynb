# Installation

Based on instructions in [Torres-Manzanera, E. xkcd: An R package for plotting XKCD graphs. Journal of Statistical Software.](https://cran.r-project.org/web/packages/xkcd/vignettes/xkcd-intro.pdf)

```
renv::init()
renv::install("extrafont")
library(extrafont)
download.file("http://simonsoftware.se/other/xkcd.ttf", dest="~/Downloads/xkcd.ttf", mode="wb")
system("mv ~/Downloads/xkcd.ttf ~/Library/Fonts")
font_import(pattern = "[X/x]kcd", prompt=FALSE, paths = "~/Library/Fonts")
```
