```
renv::init()
renv::install("extrafont")
library(extrafont)
download.file("http://simonsoftware.se/other/xkcd.ttf", dest="~/Downloads/xkcd.ttf", mode="wb")
system("mv ~/Downloads/xkcd.ttf ~/Library/Fonts")

# Make sure this is done when the renv is loaded. Otherwise you'll get many warnings while plotting
font_import(pattern = "[X/x]kcd", prompt=FALSE, paths = "~/Library/Fonts")
```

# Hypothetical age effect plots

````r
library(tidyverse)
````

````r
dat = tibble(age = seq(5, 25, 1),
        y = (age-15)^2)

xrange = range(dat$age)
yrange  = range(dat$y)

dat %>%      
      ggplot(aes(age, y))+
      geom_smooth(method = 'loess', formula = 'y ~ x')+
      labs(y = "")+
      theme(axis.text.y = element_blank(),
            axis.ticks.y = element_blank())+
      theme_xkcd()+
      xkcdaxis(xrange,yrange) 
````
